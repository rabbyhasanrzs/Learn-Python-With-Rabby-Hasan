## Python if, elif, else, and Shorthand if

These statements are used to make decisions in a program.

### 1. if Statement

Executes a block of code only if the condition is True.

Syntax
```python
if condition:
    # Code to execute
```

````python
age = 20

if age >= 18:
    print("You are an adult.")
````

### 2. if...else Statement

Executes one block if the condition is True; otherwise, executes another block.

Syntax

```python
if condition:
    # True block
else:
    # False block
```

````python
age = 16

if age >= 18:
    print("You can vote.")
else:
    print("You cannot vote.")
````

### 3. if...elif...else Statement

Checks multiple conditions in order.

Syntax

```python
if condition1:
    # Code
elif condition2:
    # Code
elif condition3:
    # Code
else:
    # Code
```

````python
marks = 75

if marks >= 80:
    print("Grade A")
elif marks >= 60:
    print("Grade B")
elif marks >= 40:
    print("Grade C")
else:
    print("Fail")
````

## 4. Nested if

An if statement inside another if statement.

````python
age = 20
has_id = True

if age >= 18:
    if has_id:
        print("Entry Allowed")
    else:
        print("ID Required")
else:
    print("You are underage.")
````

## 5. Shorthand if

A one-line if statement.

Syntax

```if condition: statement```

````python
age = 20

if age >= 18: print("Adult")
````

## 10. `pass` Statement

The `pass` statement is a **null statement**. It does **nothing** when executed.

It is used as a **placeholder** when Python requires a statement but you don't want to write any code yet.

---

## Syntax

```python
pass
```

---

### Example 1: Using `pass` with `if`

```python
age = 18

if age >= 18:
    pass

print("Program continues...")
```

**Output**

```text
Program continues...
```

---

### Example 2: Empty Function

```python
def greet():
    pass

print("Function created successfully.")
```

**Output**

```text
Function created successfully.
```

---

### Example 3: Empty Class

```python
class Student:
    pass

print("Class created successfully.")
```

**Output**

```text
Class created successfully.
```

---

### Example 4: Empty Loop

```python
for i in range(5):
    pass

print("Loop completed.")
```

**Output**

```text
Loop completed.
```

---

### Example 5: `pass` inside `if...else`

```python
number = 10

if number > 0:
    pass
else:
    print("Negative Number")

print("Done")
```

**Output**

```text
Done
```

---

### Why Use `pass`?

- ✅ To create an empty function.
- ✅ To create an empty class.
- ✅ To create an empty loop.
- ✅ To write code later without getting an error.
- ✅ Useful while developing large programs.

---

### `pass` vs `continue` vs `break`

| Statement | Meaning | Effect |
|-----------|---------|--------|
| `pass` | Do nothing | Placeholder; execution continues normally. |
| `continue` | Skip the rest of the current loop iteration | Moves to the next iteration of the loop. |
| `break` | Exit the loop immediately | Stops the loop completely. |

---

### Example

```python
print("Using pass")
for i in range(5):
    if i == 2:
        pass
    print(i)
```

**Output**

```text
Using pass
0
1
2
3
4
```

---

```python
print("Using continue")
for i in range(5):
    if i == 2:
        continue
    print(i)
```

**Output**

```text
Using continue
0
1
3
4
```

---

```python
print("Using break")
for i in range(5):
    if i == 2:
        break
    print(i)
```

**Output**

```text
Using break
0
1
```

---

### Summary

| Statement | Purpose |
|-----------|---------|
| `pass` | Placeholder; does nothing |
| `continue` | Skip the current loop iteration |
| `break` | Exit the loop immediately |

> **Easy Memory Trick**
>
> - **pass** → "Do nothing."
> - **continue** → "Skip this iteration."
> - **break** → "Stop the loop."
