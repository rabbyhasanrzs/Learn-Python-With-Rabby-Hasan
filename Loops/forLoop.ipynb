## Python Loops

A **loop** is a programming construct that **repeats a block of code** multiple times until a condition is met or all items in a collection have been processed.

Python has **2 types of loops**:

1. **`for` Loop**
2. **`while` Loop**

---

### 1. `for` Loop

## Definition

A **`for` loop** is used to iterate over a sequence (such as a list, tuple, string, dictionary, set, or range) and execute a block of code for each item.

## Syntax

```python
for variable in sequence:
    # Code to execute
```

````python
fruits = ["apple","banana","mango"];
for x in fruits:
    print(x)
````

````python
name = "Rabby Hasan";
for nam in name:
    print(nam);
````

## Loop Control Statements

Python provides **3 loop control statements**.

---

### 1. `break`

### Definition

The **`break` statement** immediately terminates the nearest enclosing loop.

### Syntax

```python
for variable in sequence:
    if condition:
        break
```

or

```python
while condition:
    if condition:
        break
```



````python
name = "RABBY HASAN"
for nam in name:
    print(nam)
    if nam==" ":
        break
````


### 2. `continue`

### Definition

The **`continue` statement** skips the remaining statements in the current iteration and moves to the next iteration.

### Syntax

```python
for variable in sequence:
    if condition:
        continue
```

or

```python
while condition:
    if condition:
        continue
```


````python
name = "RABBY HASAN"
for nam in name:
    if nam==" ":
        continue
    print(nam)
````

### 3. `pass`

### Definition

The **`pass` statement** is a placeholder that does nothing. It is used when a statement is syntactically required but no action is needed.

### Syntax

```python
for variable in sequence:
    pass
```

or

```python
while condition:
    pass
```

````python
name = "RABBY HASAN"
for nam in name:
    pass
````
