## Arithmetic Operators
Arithmetic operators are used with numeric values to perform common mathematical operations:
1. Addition
2. Subtraction
3. Multiplication	
4. Division 	
5. Modulus 	
6. Exponentiation 	
7. Floor division

### 1. Addition :


```python
x = 15
y = 4

print(x + y)
```

### 2. Subtraction

```python
x = 15
y = 4

print(f"The result is = {x-y}");
```

### 3. Multiplication

```python
x = 15
y = 4

print(f"The result is = {x*y}");
```

### 4. Division 

```python
x = 15
y = 4

print(f"The result is = {x/y}");
```

### 5. Modulus 	

```python
x = 15
y = 4

print(f"The result is = {x%y}");
```

### 6. Exponentiation 

```python
x = 15
y = 4

print(f"The result is = {x**y}");
```

### 7. Floor division

```python
x = 15
y = 4

print(f"The result is = {x//y}");
```
