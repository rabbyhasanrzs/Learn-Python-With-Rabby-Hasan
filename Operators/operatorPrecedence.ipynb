# 🐍 Python Operator Precedence Chart

> **Higher precedence operators are evaluated before lower precedence operators.**

| Precedence | Operator(s) | Description | Associativity |
|------------|-------------|-------------|---------------|
| 1 (Highest) | `()` | Parentheses (Grouping) | Left → Right |
| | `f()` | Function Call | Left → Right |
| | `x[index]` | Indexing | Left → Right |
| | `x[start:stop]` | Slicing | Left → Right |
| | `obj.attribute` | Attribute Reference | Left → Right |
| 2 | `await x` | Await Expression | Right → Left |
| 3 | `**` | Exponentiation | Right → Left |
| 4 | `+x`, `-x`, `~x` | Unary Plus, Unary Minus, Bitwise NOT | Right → Left |
| 5 | `*`, `@`, `/`, `//`, `%` | Multiplication, Matrix Multiplication, Division, Floor Division, Modulus | Left → Right |
| 6 | `+`, `-` | Addition, Subtraction | Left → Right |
| 7 | `<<`, `>>` | Bitwise Shift | Left → Right |
| 8 | `&` | Bitwise AND | Left → Right |
| 9 | `^` | Bitwise XOR | Left → Right |
| 10 | `\|` | Bitwise OR | Left → Right |
| 11 | `<`, `<=`, `>`, `>=`, `==`, `!=`, `in`, `not in`, `is`, `is not` | Comparison, Membership, Identity | Left → Right |
| 12 | `not` | Logical NOT | Right → Left |
| 13 | `and` | Logical AND | Left → Right |
| 14 | `or` | Logical OR | Left → Right |
| 15 | `if ... else` | Conditional (Ternary) Expression | Right → Left |
| 16 | `lambda` | Lambda Expression | Right → Left |
| 17 (Lowest) | `:=` | Assignment Expression (Walrus Operator) | Right → Left |
