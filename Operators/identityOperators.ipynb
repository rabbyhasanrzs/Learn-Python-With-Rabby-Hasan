## Identity Operators
Identity operators are used to compare the objects, not if they are equal, but if they are actually the same object, with the same memory location:

1. is
2. is not

### 1. is:
1. Returns True if both variables are the same object
2. The is operator returns True if both variables point to the same object:

```python
x=[1,2,3];
y=[1,2,3];
z=x

print(x is z)
print(x is y)
```

### 2. is not:
1. Returns True if both variables are not the same object
2. The is not operator returns True if both variables do not point to the same object even tow variable have same data.

```python
x = ["apple", "banana"]
y = ["apple", "banana"]

print(x is not y)
```

## Difference Between is and ==
1. is - Checks if both variables point to the same object in memory
2. == - Checks if the values of both variables are equal

```python
x = [1, 2, 3]
y = [1, 2, 3]

print(x == y)
print(x is y)
```
