## 3. Module and Pip:




### 1. What is a Module?

A **module** is simply a Python file (`.py`) that contains code such as functions, classes, and variables that you can reuse in other Python programs.

Think of a module as a **toolbox**.

For example, Python has a built-in module called `math`:

```python
import math

print(math.sqrt(25))
```

**Output**

```text
5.0
```

Here:

- `math` is the module.
- `sqrt()` is a function inside the module.

---

### Types of Modules

### 1. Built-in Modules

These come with Python automatically.

Examples:

```python
import math
import random
import os
import datetime
import statistics
```

Example:

```python
import random

print(random.randint(1, 10))
```

---

### 2. User-defined Modules

You can create your own module.

**mymath.py**

```python
def add(a, b):
    return a + b
```

Now use it in another file:

```python
import mymath

print(mymath.add(5, 7))
```

**Output**

```text
12
```

---

### 3. Third-party Modules

These are created by other developers and are **not included with Python** by default.

Examples:

- `numpy`
- `pandas`
- `opencv-python`
- `matplotlib`
- `scikit-learn`

### 2. What is pip?

**pip** stands for:

> **Pip Installs Packages**

It is Python's **package manager**.

Its job is to:

- Download packages
- Install packages
- Upgrade packages
- Uninstall packages

Think of it like:

| Programming Language | Package Manager |
|----------------------|-----------------|
| Python | pip |
| JavaScript | npm |
| Java | Maven |
| Rust | Cargo |

---

### Module vs Package

A **module** is a single Python (`.py`) file.

A **package** is a collection of related modules.

Example:

```text
numpy/
│
├── random.py
├── linalg.py
├── fft.py
└── ...
```

Together they form the **NumPy package**.

When you run:

```bash
pip install numpy
```

pip downloads and installs the entire package.

---

### How pip Works

Suppose you want to use **NumPy**.

You type:

```bash
pip install numpy
```

pip performs the following steps:

1. Connects to the **Python Package Index (PyPI)**.
2. Downloads the NumPy package.
3. Installs it into your Python environment.

After installation, you can use it:

```python
import numpy as np
```

---

```

```

## 4. Variable:

### Definition:
A variable is a named container used to store data (values) that a program can use and change later.


### Example :

Imagine you are a student going to university.Your teacher gives you three boxes.

1. 📦 Box 1 has a label: Name
2. 📦 Box 2 has a label: Roll
3. 📦 Box 3 has a label: CGPA

Now you put information into each box.

1. Name  → Rabby Hasan
2. Roll  → 221-15-5659
3. CGPA  → 3.95

In Python, these boxes are called variables.

1. name = "Rabby Hasan"
2. roll = "221-15-5659"
3. cgpa = 3.95

Here:

1. name is the variable (box name)
"Rabby Hasan" is the value inside the box
2. roll is the variable
"221-15-5659" is the value
3. cgpa is the variable
3.95 is the value

````python
name="Rabby";
roll=23;
cgpa=3.91;

print(" The name is : ",name, "Roll = ",roll, "CGPA = ",cgpa);
````

## 5. Python Data Types:

### Built-in Data type:

A **data type** tells Python **what kind of data is stored inside a variable**.

> **Definition:**  
> A **data type** is the category or type of value that a variable stores, such as text, numbers, lists, or true/false values.

---



### Why Do We Need Data Types?

Different types of data need different operations.

For example:

- We can **add numbers**.
- We can **join text**.
- We can **store multiple values in a list**.
- We can **store unique values in a set**.

Python uses data types to understand what operations are allowed.

---

### Python Built-in Data Types

| Category | Data Type | Description |
|----------|-----------|-------------|
| Text Type | `str` | Stores text (strings) |
| Numeric Types | `int`, `float`, `complex` | Stores numbers |
| Sequence Types | `list`, `tuple`, `range` | Stores ordered collections |
| Mapping Type | `dict` | Stores data as key-value pairs |
| Set Types | `set`, `frozenset` | Stores unique values |
| Boolean Type | `bool` | Stores `True` or `False` |
| Binary Types | `bytes`, `bytearray`, `memoryview` | Stores binary data |
| None Type | `NoneType` | Represents no value |

### 3. Sequence Types

Sequence means **an ordered collection of items**.



### a) List (`list`)

A list is **ordered**, **changeable**, and allows duplicate values.

```python
fruits = ["Apple", "Banana", "Orange"]

print(fruits)
```

Output

```text
['Apple', 'Banana', 'Orange']
```


````python
fruits=["apple","banana","orange"];
print(fruits);
print(type(fruits));
````

### b) Tuple (`tuple`)

A tuple is **ordered** but **cannot be changed** after creation.

```python
colors = ("Red", "Green", "Blue")

print(colors)
```

Output

```text

```

````python
fruits=("apple","banana","orange");
print(fruits);
print(type(fruits));
````

### c) Range (`range`)

````python
mobile=range(9);
print(mobile);

print(type(mobile));

print(list(mobile));
````
